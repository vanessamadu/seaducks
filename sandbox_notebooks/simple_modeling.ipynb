```python
import os
import sys
sys.path.append(os.path.abspath(os.path.join(os.path.dirname('seaducks'), '..')))
sys.path.append(os.path.abspath(os.path.join(os.path.dirname('seaducks'), '..')))
import pandas as pd
from seaducks.analysis.error_metrics.rmse_class import RMSE
from seaducks.analysis.error_metrics.sems_class import SEMs
from seaducks.analysis.error_metrics.aoems_class import AOEMs

from seaducks.models.linear_regression_model import LinearRegressionModel as lr
from seaducks.models.baseline_model import BathtubModel
import matplotlib.pyplot as plt
# configure plots
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "Helvetica"
})

plt.style.use('ggplot')
```

```python
drifter_dataset = pd.read_hdf('../data/filtered_nao_drifters_with_sst_gradient.h5')
drifter_dataset['day_of_year'] = drifter_dataset['time'].apply(lambda t : t.timetuple().tm_yday)
```

```python
import numpy as np
rmse_values = []
speed_mae_values = []
mean_absolute_angle_offsets = []

covs = ['u_av','v_av','lat','lon','day_of_year','Wx','Wy','Tx','Ty','sst_x_derivative','sst_y_derivative']
lr_model = lr(drifter_dataset,covs)

predictions = lr_model.predictions
observations = lr_model.observations

err = RMSE(predictions,observations).error
speed_mae = SEMs(predictions,observations).mae_speed.error
aome_model = AOEMs(predictions,observations)

maao = aome_model.maao_all.error

rmse_values.append(err)
speed_mae_values.append(speed_mae)
mean_absolute_angle_offsets.append(maao)

for ii,cov in enumerate(covs):
    new_covs = np.array([val for val in covs if val !=cov])
    lr_model = lr(drifter_dataset,new_covs)

    predictions = lr_model.predictions
    observations = lr_model.observations

    err = RMSE(predictions,observations).error
    speed_mae = SEMs(predictions,observations).mae_speed.error
    aome_model = AOEMs(predictions,observations)

    maao = aome_model.maao_all.error

    rmse_values.append(err)
    speed_mae_values.append(speed_mae)
    mean_absolute_angle_offsets.append(maao)
```
